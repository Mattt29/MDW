```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Obtenir les noms de feuilles
annee = list(xls.keys())
annee=['1938-2007', '2008-2009', '2009-2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
filename = 'VilleMTP_MTP_Tournages_renom.xls'
xls = pd.read_excel(filename, sheet_name=None)
files1=pd.DataFrame()
files2=pd.DataFrame()

for year in annee:
  data = pd.read_excel('VilleMTP_MTP_Tournages_renom.xls', sheet_name=year)
  data.drop(data.loc[data.iloc[:,0] == 'Type'].index, inplace=True)
  data.iloc[:, 0] = data.iloc[:, 0].fillna(method='ffill')
  data.rename(columns={'Nom du projet':'Titre'},inplace=True)
  data['Genre']=pd.Series(index=data.index)
  data['Acteurs']=pd.Series(index=data.index)
  data['Tournage à Mntp']=pd.Series(index=data.index)
  data['Sortie en France']=pd.Series(index=data.index)
  data['Prix et nominations']=pd.Series(index=data.index)
  data=data.assign(annee=year)
  files1=pd.concat([files1,data])

nouvelle_annee=['2020','2021']
for year in nouvelle_annee:
  data = pd.read_excel('VilleMTP_MTP_Tournages_renom.xls', sheet_name=year, header=1)
  data.rename(columns={'Lieux filmés':'Lieux'},inplace=True)
  data.rename(columns={'Genre':'Type','Producteur et Société de production':'Production'},inplace=True)
  data['Unnamed: 0']=pd.Series(index=data.index)
  data['Nombre de jours']=pd.Series(index=data.index)
  data['Dates Début tournage']=pd.Series(index=data.index)
  data['Dates Fin tournage']=pd.Series(index=data.index)
  data=data.assign(annee=year)
  files2=pd.concat([files2,data])
```

```python
json_data1 = files1.to_json(orient='records', force_ascii=False)
json_data2 = files2.to_json(orient='records', force_ascii=False)
```

```python
import json

data1 = json.loads(json_data1)
data2 = json.loads(json_data2)

data1.extend(data2)

json_data_concat = json.dumps(data1, ensure_ascii=False)
```
